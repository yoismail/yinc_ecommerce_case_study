### Importing Neccessary Libraries 

```python
import pandas as pd
import numpy as np
```

### Extraction Layer

```python
path = r'dataset\raw_data\yinc_ecommerce.csv'
yinc_df = pd.read_csv(path)
```

```python
yinc_df.head()
yinc_df.info()
```

```python
#Cleaning Columns
yinc_df.columns = (
    yinc_df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
yinc_df.columns
```

```python
yinc_df.isnull().sum()
```

### Data Cleaning and Transformation

```python
# Drop Missing Values 
yinc_df = yinc_df.dropna(subset=['order_id', 'customer_id', 'customer_name', 'state'])
```

```python
# Clean order_date datatype (Str - DateTime Object)
yinc_df['order_date'] = pd.to_datetime(yinc_df['order_date'], format = "%d/%m/%Y %H:%M", errors='coerce')

# Sanity Check
yinc_df['order_date'].head()
```

```python
yinc_df.isnull().sum()
```

### Create Tables

```python
# Customer Table
customer_df = yinc_df[['customer_id','customer_name', 'email', 'phone_number']].drop_duplicates().reset_index(drop=True).copy()
customer_df.head()
```

```python
# Delete and create a new Email Address Column to Customer Table
del[customer_df["email"]]
customer_df['email_address'] = (
    customer_df['customer_name']
    .str.lower()
    .str.strip()
    .str.replace(" ","_")
    + '@'
    + customer_df['customer_id']
    + '.com'
    )
customer_df.head()
```

```python
# Product Table
product_df = yinc_df[['product_id', 'product_name', 'brand', 'category', 'price' ]].drop_duplicates().reset_index(drop=True).copy()
product_df.head()
```

```python
# Shipping Table
shipping_df = yinc_df[['customer_id', 'shipping_address','city', 'state', 'country', 'postal_code']].drop_duplicates().reset_index(drop=True).copy()

shipping_df.index.name = 'shipping_id'
shipping_df = shipping_df.reset_index()

shipping_df.head()
```

```python
# Order Table
order_df = yinc_df[['order_id', 'customer_id','product_id', 'quantity', 'total_price', 'order_date']].drop_duplicates().reset_index(drop=True).copy()
order_df.head()
```

```python
# Payment Method Table
payment_method_df = yinc_df[['order_id','payment_method','transaction_status']].drop_duplicates().reset_index(drop=True).copy()
payment_method_df.head()
```

### Save Tables to CSV in Cleaned Data Folder


```python
# Save Table to CSV
customer_df.to_csv(r'dataset/cleaned_data/customer.csv', index=False)
product_df.to_csv(r'dataset/cleaned_data/product.csv', index=False)
shipping_df.to_csv(r'dataset/cleaned_data/shipping.csv', index=False)
order_df.to_csv(r'dataset/cleaned_data/order.csv', index=False)
payment_method_df.to_csv(r'dataset/cleaned_data/payment_method.csv', index=False)
```

### Creating SQL Table Queries


```python
def create_tables():
    conn = get_db_connection()
    cursor = conn.cursor()
    create_table_query = '''
                            CREATE SCHEMA IF NOT EXISTS yinc;
                            
                            DROP TABLE IF EXISTS yinc.customer CASCADE;
                            DROP TABLE IF EXISTS yinc.product  CASCADE;
                            DROP TABLE IF EXISTS yinc.shipping CASCADE;
                            DROP TABLE IF EXISTS yinc.order CASCADE;
                            DROP TABLE IF EXISTS yinc.payment_method CASCADE;
                            
                            CREATE TABLE IF NOT EXISTS yinc.customer (
                                customer_id    UUID PRIMARY KEY,
                                customer_name  TEXT,
                                phone_number   TEXT,
                                email_address  TEXT
                            );
                            
                             CREATE TABLE IF NOT EXISTS yinc.product (
                                product_id     UUID PRIMARY KEY,
                                product_name   TEXT,
                                brand          TEXT,
                                category       TEXT,
                                price          FLOAT
                            );
                            
                            CREATE TABLE IF NOT EXISTS yinc.shipping (
                                shipping_id              INT PRIMARY KEY,
                                customer_id              UUID,
                                shipping_address         TEXT,
                                city                     TEXT,
                                state                    TEXT,
                                country                  TEXT,
                                postal_code              INT,
                                FOREIGN KEY (customer_id) REFERENCES yinc.customer(customer_id)    
                            );
                            
                              CREATE TABLE IF NOT EXISTS yinc.order (
                                order_id                 UUID PRIMARY KEY,
                                customer_id              UUID,
                                product_id               UUID,
                                quantity                 INT,
                                total_price              NUMERIC(10,2),
                                order_date               DATE,
                                FOREIGN KEY (customer_id) REFERENCES yinc.customer(customer_id), 
                                FOREIGN KEY (product_id) REFERENCES yinc.product(product_id)   
                            );
                            
                             CREATE TABLE IF NOT EXISTS yinc.payment_method (
                                order_id                 UUID, 
                                payment_method           TEXT,
                                transaction_status       TEXT,
                                FOREIGN KEY (order_id) REFERENCES yinc.order(order_id)    
                            );'''
    
    try:
        cursor.execute(create_table_query)
        conn.commit()

    except Exception as e:
        print("Error:", e)
        conn.rollback()

    finally:
        cursor.close()
        conn.close()
```

```python
print("Creating Table")
create_tables()
print("Table Query Created...........")
```

### Loading CSV Files to Postgres

```python
import csv
def load_data_from_csv(csv_path):
    conn = get_db_connection()
    cursor = conn.cursor()
    with open(csv_path, "r") as file:
        reader = csv.reader(file)
        next(reader)
        for row in reader:
            cursor.execute('''
                           INSERT INTO yinc.customer (customer_id, customer_name, phone_number, email_address)
                           VALUES (%s, %s, %s, %s);''',
                           row
                        )
    conn.commit()
    cursor.close()
    conn.close()

csv_file_path = r'dataset\cleaned_data\customer.csv'

load_data_from_csv(csv_file_path)
```
